```python
for index in range(len(raw_data['Gender'])):
    if raw_data['Gender'].loc[index]:
        raw_data['Gender'].loc[index] = 'Male'
    else:
        raw_data['Gender'].loc[index] = 'Female'

raw_data
```

```python
for index in range(len(raw_data['family_history_with_overweight'])):
    if raw_data['family_history_with_overweight'].loc[index]:
        raw_data['family_history_with_overweight'].loc[index] = 'Yes'
    else:
        raw_data['family_history_with_overweight'].loc[index] = 'No'
        
for index in range(len(raw_data['SMOKE'])):
    if raw_data['SMOKE'].loc[index]:
        raw_data['SMOKE'].loc[index] = 'Yes'
    else:
        raw_data['SMOKE'].loc[index] = 'No'      

for index in range(len(raw_data['Automobile'])):
    if raw_data['Automobile'].loc[index]:
        raw_data['Automobile'].loc[index] = 'Yes'
    else:
        raw_data['Automobile'].loc[index] = 'No'
        
for index in range(len(raw_data['Bike'])):
    if raw_data['Bike'].loc[index]:
        raw_data['Bike'].loc[index] = 'Yes'
    else:
        raw_data['Bike'].loc[index] = 'No'      

for index in range(len(raw_data['Motorbike'])):
    if raw_data['Motorbike'].loc[index]:
        raw_data['Motorbike'].loc[index] = 'Yes'
    else:
        raw_data['Motorbike'].loc[index] = 'No' 

for index in range(len(raw_data['Public_Transportation'])):
    if raw_data['Public_Transportation'].loc[index]:
        raw_data['Public_Transportation'].loc[index] = 'Yes'
    else:
        raw_data['Public_Transportation'].loc[index] = 'No'
        
for index in range(len(raw_data['Walking'])):
    if raw_data['Walking'].loc[index]:
        raw_data['Walking'].loc[index] = 'Yes'
    else:
        raw_data['Walking'].loc[index] = 'No'
        
for index in range(len(raw_data['NObeyesdad'])):
    if raw_data['NObeyesdad'].loc[index]:
        raw_data['NObeyesdad'].loc[index] = 'Yes'
    else:
        raw_data['NObeyesdad'].loc[index] = 'No'

for index in range(len(raw_data['CAEC'])):
    if raw_data['CAEC'].loc[index]:
        raw_data['CAEC'].loc[index] = 'Yes'
    else:
        raw_data['CAEC'].loc[index] = 'No'
        
for index in range(len(raw_data['FAVC'])):
    if raw_data['FAVC'].loc[index]:
        raw_data['FAVC'].loc[index] = 'Yes'
    else:
        raw_data['FAVC'].loc[index] = 'No'
        
for index in range(len(raw_data['CH2O'])):
    if raw_data['CH2O'].loc[index]:
        raw_data['CH2O'].loc[index] = 'Yes'
    else:
        raw_data['CH2O'].loc[index] = 'No'
        
for index in range(len(raw_data['SCC'])):
    if raw_data['SCC'].loc[index]:
        raw_data['SCC'].loc[index] = 'Yes'
    else:
        raw_data['SCC'].loc[index] = 'No'
```

```python
raw_data.rename(columns={'NObeyesdad':'Obese'}, inplace=True)
raw_data.rename(columns={'FAVC':'frequently_consumed_high_calorie'}, inplace=True)
raw_data.rename(columns={'FCVC':'frequency_of_consumpton_of_vegetables'}, inplace=True)
raw_data.rename(columns={'NCP':'nursing_care_plan'}, inplace=True)
raw_data.rename(columns={'SCC':'calories_consumption_monitoring'}, inplace=True)
raw_data.rename(columns={'CH2O':'amount_of_water_consumption'}, inplace=True)

# nursing care plan is defined as assess weight and blood pressure once or twice a week!
# CAEC; CAE Comorbidities ; children with newly diagnosed CAE have higher chances oof overweight and obesity
```

# Questions and Answers

## 1. How many people in the dataset are obese and not?

```python
raw_data['Obese'].value_counts()
```

```python
print('''Number of obese people: {}
Number of people not obese: {}'''.format(raw_data['Obese'].value_counts()[1], raw_data['Obese'].value_counts()[0]))
```

## 2. How many Obese people has a family history of overweight?

```python
a = raw_data[(raw_data['Obese'] == 'Yes') & (raw_data['family_history_with_overweight'] == 'Yes')].shape
a
```

```python
print('Number of obese people with family history of overweight: {}'.format(a[0]))
```

## 3. How many obese people smoke?

```python
b = raw_data[(raw_data['Obese'] == 'Yes') & (raw_data['SMOKE'] == 'Yes')].shape
b
```

```python
print('Number of obese people that smoke: {}'.format(b[0]))
```

## 4. What is the relationship of those that smoke and obesity?

```python
print('''
Number of obese people that smoke: {}
Number of Obese people that don't smoke: {}

Answer: Smoking doesn't make one obese neither those it contribute to weight gain'''.format(b[0], (raw_data['Obese'].value_counts()[1] - b[0])))
```

## 5. what's the number of obese people who walk and not?

```python
c = raw_data[(raw_data['Obese'] == 'Yes') & (raw_data['Walking'] == 'Yes')].shape
d = raw_data[(raw_data['Obese'] == 'Yes') & (raw_data['Walking'] == 'No')].shape
c,d
```

```python
print('''
Number of obese people that walk: {}
Number of obese people that don't walk: {}
'''.format(c[0], d[0]))
```

# Inferences and Conclusions

> 1. Both Males and Females are equally prone to obesity, indicating that gender does not significantly affect obesity rates.

> 2. Individuals who do not engage in regular physical activity have a higher likelihood of gaining weight. Those who use public transport or engage in regular exercise(e.g walking or cycling) are less likely to be obese.

> 3. There is a strong correlation between family history and obesity. Inividuals with obese family members have a higher risk of becoming obese themselves.

These conclusions highlight the importance of lifestyle choices in managing obesity. Regular screening and healthy lifestyle choices are crucial in mitigating risks associated with obesity.
